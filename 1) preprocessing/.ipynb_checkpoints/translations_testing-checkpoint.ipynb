```python
from langdetect import detect_langs
from validator_collection import validators, checkers
import unidecode, re
import pandas as pd
```

```python
def detect_language(row, text_columns):
    global corrupt_ids
    error = []
    for col in text_columns:
        try:
            text = unicode(row[col], "utf-8")
            if len(text):
                if checkers.is_url(text):
                    continue
                if not re.match(r'^[_\W]+$', text):
                    translation = detect_langs(text)
                    for item in translation:
                        if item.lang == "de" and item.prob > 0.5:
                            error.append(col)
        except:
            corrupt_ids.append(row['id'])
    if len(error):
        row['translation_issues'] = error
        row['translation_status'] = 'error'
    return row
```

```python
corrupt_ids = []
df = pd.read_csv('../datasets/mantis_en_issuess_v6.csv')
df = df.fillna('')
text_columns = [u'summary', u'description', u'additional_information', u'steps_to_reproduce', u'bug_notes']
```

```python
temp = df.apply(lambda x: detect_language(x, text_columns), axis=1)
```

```python
temp.to_csv('../datasets/mantis_en_issuess_v6.csv', index=False)
```
