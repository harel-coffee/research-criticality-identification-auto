```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
bugs_vector = pd.read_csv('../datasets/mantis_bugs_vector.csv', index_col='id')
bugs_vector['pm_ticket'] = bugs_vector['pm_ticket'].fillna(False)
```

```python
senti_vector = bugs_vector[[u'extremely_negative', u'extremely_positive', u'negative', u'neutral', u'positive']].copy()
```

```python
continuous_cols = bugs_vector[[u'extremely_negative', u'extremely_positive', u'negative', u'neutral', u'positive']]
discrete_cols = bugs_vector[['pm_ticket', 'priority', 'severity']]
```

```python
d_c = {}
fig = plt.figure(figsize = (12, 10))
title = fig.suptitle("Discrete Column Frequency", fontsize=14)
for index, col in enumerate(discrete_cols.columns):
    ax = fig.add_subplot(2,2, index+1)
    d_c[col] = discrete_cols[col].value_counts()
    ax = d_c[col].sort_index().plot('bar', ax=ax, color='steelblue', edgecolor='black', rot=False)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
```

```python
continuous_cols.replace(0, np.nan).plot.kde(subplots=True, layout=(3,2), figsize=(15, 10), sharex=False, ylim=(0, 1.5), xlim = (0, 15))
```

```python
f, ax = plt.subplots(figsize=(10, 6))
corr = bugs_vector.corr()
hm = sns.heatmap(corr, annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('MANTIS Attributes Correlation Heatmap', fontsize=14)
```

```python
pp = sns.pairplot(bugs_vector, size=1.8, aspect=1.8,
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Mantis Attributes Pairwise Plots', fontsize=14)
```

# Sentiment Classification based clustering

```python
preprocess_senti = make_column_transformer(
    (StandardScaler(), 
     [u'extremely_negative', u'extremely_positive', u'negative', u'neutral', u'positive'])
)
```

```python
X_senti = preprocess_senti.fit_transform(senti_vector)
```

```python
pca = sklearnPCA(n_components=2)
X_senti_pca = pca.fit_transform(X_senti)
```

```python
transformed = pd.DataFrame(X_senti_pca)
```

```python
plt.scatter(transformed[0], transformed[1])
```

```python
del_ind = transformed[(transformed[0]>12) | (transformed[1]>10) | (transformed[1]<-10)].index
```

```python
transformed.drop(del_ind, axis=0, inplace = True)
```

```python
plt.scatter(transformed[0], transformed[1])
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=750).fit(X_senti_pca)
transformed['cluster_pca_db'] = dbscan.labels_
```

```python
plt.scatter(X_senti_pca[:,0], X_senti_pca[:,1], c=dbscan.labels_)
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X_senti_pca)
labels = gmm.predict(X_senti_pca)
transformed['cluster_pca_em'] = labels
```

```python
plt.scatter(transformed[0], transformed[1], c=labels)
```

```python
print transformed['cluster_pca_db'].value_counts()
print transformed['cluster_pca_em'].value_counts()
```

```python
#lda = LDA(n_components=2) #2-dimensional LDA
#lda_transformed = pd.DataFrame(lda.fit_transform(X_senti))
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X_senti)
senti_vector['cluster'] = kmeans.labels_
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=750).fit(X_senti)
senti_vector['cluster_dbscan'] = dbscan.labels_
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X_senti)
labels = gmm.predict(X_senti)
senti_vector['cluster_GMM'] = labels
```

```python
print senti_vector['cluster'].value_counts()
print senti_vector['cluster_dbscan'].value_counts()
print senti_vector['cluster_GMM'].value_counts()
```

# Clustering on Sentiment Classification features and the meta data

```python
preprocess = make_column_transformer(
    (StandardScaler(), 
     [u'extremely_negative', u'extremely_positive', u'negative', u'neutral', u'positive', u'priority', u'severity']),
    (OneHotEncoder(), ['pm_ticket'])
)
```

```python
X = preprocess.fit_transform(bugs_vector)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

```python
len(kmeans.labels_)
```

```python
bugs_vector['cluster'] = kmeans.labels_
```

```python
bugs_vector['cluster'].value_counts()
```

```python
bugs_vector[bugs_vector['cluster']==1]
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=750).fit(X)
```

```python
len(dbscan.labels_)
```

```python
bugs_vector['cluster_dbscan'] = dbscan.labels_
```

```python
bugs_vector['cluster_dbscan'].value_counts()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X)
labels = gmm.predict(X)
bugs_vector['cluster_GMM'] = labels
bugs_vector['cluster_GMM'].value_counts()
```

```python
pp = sns.pairplot(bugs_vector, size=1.8, aspect=1.8,
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Mantis Attributes Pairwise Plots', fontsize=14)
```

```python
# Visualizing 6-D mix data using scatter charts
# leveraging the concepts of hue, facets and size
g = sns.FacetGrid(bugs_vector, row='priority', col="severity", hue='cluster_GMM', size=4)
g.map(plt.scatter,  "extremely_negative", "extremely_positive", alpha=0.5, 
      edgecolor='k', linewidth=0.5, s=bugs_vector['neutral']*2)
fig = g.fig 
fig.set_size_inches(18, 14)
fig.subplots_adjust(top=0.85, wspace=0.3)
fig.suptitle('priority - severity - extremely_negative - extremely_positive - neutral - pm_ticket', fontsize=14)
l = g.add_legend(title='Mantis Attributes Pairwise Plots')
```
